```python
#viridis
renkler = ['#4b4279',"#3c6682", '#2f837f', '#45a778', '#8fc456'] 
```

```python
# Veriler
#kategoriler = ['Yüksek Lisans', 'Doktora']
#degerler = [103, 9]

# Veriler
#kategoriler = ['Türkçe', 'İngilizce']
#degerler = [55, 57]

# Veriler
kategoriler = ['2020', '2021', '2022', '2023', '2024']
degerler = [15, 18, 35, 21, 25]
```

```python
# Pasta grafik oluşturma
plt.figure(figsize=(8, 6))
plt.pie(degerler, 
        labels=kategoriler, 
        autopct='%1.1f%%',
        startangle=90,
        colors=renkler,
        textprops={'fontsize': 17})  # Yüzde yazı boyutu

# Başlık ve etiket boyutlarını büyütme
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=kategoriler, y=degerler, palette="viridis")
plt.xlabel('Yıllar', fontsize=18)
plt.ylabel('Tez Sayısı', fontsize=18)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
#sns.despine()  # Çerçeveyi sadeleştirir
plt.show()
```

```python
# Grafik oluşturma
plt.figure(figsize=(10, 6))
plt.bar(kategoriler, degerler, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Örnek Sütun Grafik', fontsize=16)
plt.xlabel('Kategoriler', fontsize=14)
plt.ylabel('Değerler', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.show()
```
